# Concatenate the dataframes of each month and remove any row that has NaN values, then export the 1st Quater dataframe as csv.

```python
df_all =  pd.concat([df6, df7, df8, df9, df10, df11, df12], ignore_index=True)
df_all.tail()
```

```python
df_all['gender']
```

```python
df_all['Gender'] = df_all['gender'].replace([0,1,2],['unknown',"Male","Female"])
df_all = df_all.drop('gender',1)
df_all["distance"] = ''
df_all.tail()
```

```python
import geopy.distance
```

```python
from geopy.distance import great_circle
```

```python
geopy.distance.geodesic((40.72,-74.03),(40.72,-74.05)).km
```

```python
def get_gdistance(row):
    try:
        return(geopy.distance.geodesic((row['start station latitude'], row['start station longitude']), (row['end station latitude'], row['end station longitude'])).km)
    except:
        return("BROKEN")
```

```python
df_all['distances_km'] = df_all.apply(get_gdistance, axis = 1)
```
